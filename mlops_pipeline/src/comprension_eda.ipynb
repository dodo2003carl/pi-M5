# 2. Comprensión y Análisis Exploratorio de Datos (EDA)
Análisis detallado de la base de datos para identificar patrones y limpiar nulos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_excel('../Base_de_datos.xlsx')

## 1. Exploración Inicial
print("Caracterización de los datos:")
display(df.info())

## 2. Unificación de Nulos
df.replace([' ', 'N/A', 'None', 'nan'], np.nan, inplace=True)
print("\nConteo de nulos por columna:")
print(df.isnull().sum())
```

```python
## 3. Análisis Univariado
print("Descripción estadística básica:")
display(df.describe(include='all'))

# Histogramas para variables numéricas
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Distribución de Variables Numéricas')
plt.tight_layout()
plt.show()
```

```python
## 4. Análisis Multivariado
plt.figure(figsize=(10, 8))
correlation = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```
